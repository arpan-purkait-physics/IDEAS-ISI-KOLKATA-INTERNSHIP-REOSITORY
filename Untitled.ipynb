```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("GWB1223.csv")
print(df.head())
grids = df[['Latitude', 'Longitude']].drop_duplicates()
print("Total grids: ",len(grids))
df = df.loc[:,~df.columns.duplicated()] # Duplicated Data cleaning
df['Date'] = pd.to_datetime(df['Date']) # Date time format
print("Missing values: \n",df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
lat, lon = 21.5, 85.5
df_grid = df[(df['Latitude'] == lat) & (df['Longitude'] == lon)].copy()
df_grid = df_grid.sort_values('Date')
print(len(df_grid))
```

```python
train = df_grid[df_grid['Date'].dt.year < 2023]  
test  = df_grid[df_grid['Date'].dt.year == 2020]
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA (try p=5, d=1, q=0 first as a start)
model = ARIMA(train['Tmax'], order=(5, 1, 0))
model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(test['Date'], test['Tmax'], label='Actual Tmax', color='blue')
plt.plot(test['Date'], test['Rain'], label='Actual Tmax', color='blue')
plt.plot(test['Date'], forecast, label='Predicted Tmax', color='red', linestyle='--')
plt.title(f'Tmax Forecast vs Actual at Grid ({lat}, {lon})')
plt.xlabel('Date')
plt.ylabel('Tmax')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example for one grid point
df_grid = df[(df['Latitude'] == 21.5) & (df['Longitude'] == 85.5)]
df_grid = df_grid.set_index('Date').asfreq('D')
df_grid = df_grid[['Tmax']]  # or Tmean
```

```python
train = df_grid.loc['2012-01-01':'2021-12-31']
test = df_grid.loc['2022-01-01':'2022-12-31']
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(train['Tmax'])
plot_pacf(train['Tmax'])
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train['Tmax'], order=(1,1,1), seasonal_order=(1,1,1,365))
result = model.fit()
```

```python
pred = result.predict(start=test.index[0], end=test.index[-1])
plt.plot(train.index, train['Tmax'], label='Train')
plt.plot(test.index, test['Tmax'], label='Actual')
plt.plot(pred.index, pred, label='Predicted')
plt.legend()
plt.show()
```
